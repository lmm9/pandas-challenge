```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load, encoding="utf-8")
student_data = pd.read_csv(student_data_to_load, encoding="utf-8")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

```python
#Calculate the total number of schools
total_schools=len(school_data_complete["school_name"].value_counts())
total_schools
```

```python
#Calculate the total number of students
total_students=len(school_data_complete["Student ID"].value_counts())
total_students
```

```python
#Calculate the total budget
total_budget=school_data["budget"].sum()
total_budget
```

```python
#Calculate the average math score
avg_math=school_data_complete["math_score"].sum()/len(school_data_complete["Student ID"].value_counts())
avg_math
```

```python
#Calculate the average reading score
avg_read=school_data_complete["reading_score"].sum()/len(school_data_complete["Student ID"].value_counts())
avg_read
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
school_data_complete["math_pass"]=(school_data_complete["math_score"] >=70)
math_perc=100*len(school_data_complete.loc[school_data_complete["math_pass"]==True])/total_students

#Calculate the percentage of students with a passing reading score (70 or greater)
school_data_complete["read_pass"]=(school_data_complete["reading_score"] >=70)
read_perc=100*len(school_data_complete.loc[school_data_complete["read_pass"]==True])/total_students

#Calculate the percentage of students who passed math and reading (% Overall Passing)
school_data_complete["overall_pass"]=(school_data_complete["reading_score"] >=70)&(school_data_complete["math_score"] >=70)
overall_perc=100*len(school_data_complete.loc[school_data_complete["overall_pass"]==True])/total_students

#Display new complete Dataframe
school_data_complete.head()
```

```python
#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame([{"Total Schools":total_schools,
                                   "Total Students":total_students,
                                   "Total Budget":total_budget,
                                   "Average Math Score":avg_math,
                                   "Average Reading Score":avg_read,
                                   "% Passing Math": math_perc,
                                   "% Passing Reading": read_perc,
                                   "% Overall Passing": overall_perc}
                                  ])
                            
district_summary_df.head()
```

```python
#Give the displayed data cleaner formatting
district_summary_df["Total Schools"] = district_summary_df["Total Schools"].map("{:,}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.2f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.2f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.2f}".format)
district_summary_df.head()
```

```python
# Define the bins and groups
bins1=[0, 600, 650, 700]
groups1=["<$600", "$600-650", "$650+"] 
```

```python
#Scores by Spending Per Student Dataframe
spending_group_df=school_spending_df.groupby("Spending Per Student")
spending_group_df.mean()
```
